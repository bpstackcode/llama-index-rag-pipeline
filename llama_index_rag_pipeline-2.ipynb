```python
import os
os.environ["OPENAI_API_KEY"] = "your-api-key-here"
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.core.schema import Document

# Create sample documents directly in code
documents = [
    Document(text="A driver with more than 2 at-fault accidents in the past 3 years is considered high risk."),
    Document(text="Vehicles older than 15 years may require additional inspection before coverage is approved."),
    Document(text="Drivers under 25 years old are subject to a youth surcharge on premiums."),
    Document(text="A credit score below 600 may result in higher premium rates or denial of coverage."),
]

print("Building vector index...")
index = VectorStoreIndex.from_documents(documents)

query_engine = index.as_query_engine()
response = query_engine.query("What makes a driver high risk?")
print(f"Response: {response}")
```
